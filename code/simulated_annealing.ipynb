```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt


# Define the Rastrigin function
def rastrigin(x, y, A=10):
    return -(
        2 * A
        + (x**2 - A * np.cos(2 * np.pi * x))
        + (y**2 - A * np.cos(2 * np.pi * y))
    )


# Define the function to plot
def func_to_plot(x, y, T_t):
    h_theta = rastrigin(x, y)
    return np.exp(h_theta / T_t)


# Generate x, y values
x_values = np.linspace(-5.12, 5.12, 100)
y_values = np.linspace(-5.12, 5.12, 100)
x_mesh, y_mesh = np.meshgrid(x_values, y_values)

# Different T_t values
T_t_values = [10, 5, 1, 0.1]


# Create the 2x2 plot with reordered T_t values and adjusted layout to prevent cropping
fig, axes = plt.subplots(2, 2, figsize=(14, 14), subplot_kw={"projection": "3d"})
fig.suptitle(
    "3D Plots of $\\exp \\left(\\frac{h(\\theta)}{T_t}\\right)$ for Different $T_t$ Values"
)

for ax, T_t in zip(axes.flatten(), T_t_values):
    z_values = func_to_plot(x_mesh, y_mesh, T_t)
    ax.plot_surface(x_mesh, y_mesh, z_values, cmap="viridis")
    ax.set_title(f"$T_t = {T_t}$")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    # ax.set_zlabel("$\\exp \\left(\\frac{h(\\theta)}{T_t}\\right)$")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the function to plot
def func(T_t, delta_h=-0.1):
    return np.minimum(np.exp(delta_h / T_t), 1)


# Generate T_t values from 10 to 0.1
T_t_values = np.linspace(10, 0.1, 500)

# Calculate the new function values
func_values_negative_delta_h = func(T_t_values)

# Create the updated plot
plt.figure(figsize=(10, 6))
plt.plot(T_t_values, func_values_negative_delta_h)
plt.xlabel("$T_t$")
plt.ylabel(r"$\rho(T_t)$")  # Corrected LaTeX
plt.title(r"$\rho(T_t)$ vs. $T_t$")  # Corrected LaTeX
plt.xlim(10, 0.1)  # Reverse the x-axis to go from 10 to 0.1
plt.grid(True)
plt.show()
```
